```python
#Rolling Average
window_size=15
df['Rolling_Average']=df['Close'].rolling(window=window_size).mean()
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], color='Green', label='Close Price')
plt.plot(df['Date'], df['Rolling_Average'], color='red', label=f'{window_size}-Day Rolling Average')
plt.title('Ethereum Price and Rolling Average Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate short-term (fast) and long-term (slow) moving averages
short_window = 50  # Short-term moving average window
long_window = 200  # Long-term moving average window
df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
df['Long_MA'] = df['Close'].rolling(window=long_window).mean()
```

```python
# Generate buy/sell signals based on moving average crossovers
df['Signal'] = 0  # Initialize signal column
# When short-term MA crosses above long-term MA, generate buy signal (1)
df.loc[df['Short_MA'] > df['Long_MA'], 'Signal'] = 1
```

```python
# When short-term MA crosses below long-term MA, generate sell signal (-1)
df.loc[df['Short_MA'] < df['Long_MA'], 'Signal'] = -1
```

```python
# Plot Ethereum prices and moving averages
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], color='blue', label='Close Price')
plt.plot(df['Date'], df['Short_MA'], color='red', label=f'Short-term MA ({short_window} days)')
plt.plot(df['Date'], df['Long_MA'], color='green', label=f'Long-term MA ({long_window} days)')
```

```python
# Plot buy signals
plt.plot(df[df['Signal'] == 1]['Date'], df[df['Signal'] == 1]['Close'], '^', markersize=10, color='g', lw=0, label='Buy Signal')
```

```python
# Plot sell signals
plt.plot(df[df['Signal'] == -1]['Date'], df[df['Signal'] == -1]['Close'], 'v', markersize=10, color='r', lw=0, label='Sell Signal')
```

```python
# Plot sell signals
plt.plot(df[df['Signal'] == -1]['Date'], df[df['Signal'] == -1]['Close'], 'v', markersize=10, color='r', lw=0, label='Sell Signal')

plt.title('Moving Average Crossover Strategy')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the cumulative return
weekly_summary['Cumulative Return'] = (1 + weekly_summary['Volume']).cumprod()
```

```python
# Create the bar plot for cumulative return
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(weekly_summary['Date'], weekly_summary['Cumulative Return'], color='Green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Cumulative Return', fontsize=14)
plt.title('Cumulative Return', fontsize=14)
plt.grid()
plt.show()
```
